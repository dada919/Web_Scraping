- Obtenir les derniers articles du Monde (**)

```python
!pip install unidecode
import requests
from bs4 import BeautifulSoup
import pandas as pd
from unidecode import unidecode
import os
```

```python
url = "https://www.lemonde.fr/actualite-en-continu/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("section", class_="teaser")
for article in articles:
    title = article.find("h3", class_="teaser__title").text.strip()
    chapeau = article.find("p", class_="teaser__desc").text.strip()
    print("Titre:", title)
    print("Chapeau:", chapeau)
    print()
```

```python
if not os.path.exists("images"):
    os.makedirs("images")

url = "https://www.lemonde.fr/actualite-en-continu/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("section", class_="teaser")
for i, article in enumerate(articles, 1):
    title = article.find("h3", class_="teaser__title").text.strip()
    chapeau = article.find("p", class_="teaser__desc").text.strip()
    image_url = article.find("figure", class_="teaser__figure").find("img")["data-src"]
    
    image_response = requests.get(image_url)
    image_extension = image_url.split(".")[-1]
    image_filename = f"images/article_{i}.{image_extension}"
    with open(image_filename, "wb") as f:
        f.write(image_response.content)
    
    print("Titre:", title)
    print("Chapeau:", chapeau)
    print("Image enregistrée:", image_filename)
    print()
```

- Le beau temps fait-il le bonheur ? (**)

```python
depression_url = "https://en.wikipedia.org/wiki/Epidemiology_of_depression"
sunshine_url = "https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration"
```

```python
def scrape_table(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    table = soup.find("table")
    df = pd.read_html(str(table))[0]
    return df

depression_df = scrape_table(depression_url)
sunshine_df = scrape_table(sunshine_url)

print("Tableau sur la dépression :")
print(depression_df.head())
print("\nTableau sur les heures d'ensoleillement :")
print(sunshine_df.head())
```

```python
def clean_sunshine_data(df):
    df.columns = ["1990", "2017"]
    return df

def clean_depression_data(df):
    df = df[["Country"]]
    return df

def calculate_average_sunshine(df):
    df["Average_Sunshine"] = df.mean(axis=1)
    average_sunshine = df[["Country", "Average_Sunshine"]]
    return average_sunshine

cleaned_sunshine_df = clean_sunshine_data(sunshine_df)
cleaned_depression_df = clean_depression_data(depression_df)
average_sunshine_per_country = calculate_average_sunshine(cleaned_sunshine_df)

print("Moyennes annuelles d'ensoleillement par pays :")
print(average_sunshine_per_country)
```

- Le meilleur kebab du coin

```python
url = 'https://www.kebab-frites.com/meilleur-kebab/paris-d54.html'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    restaurants = soup.find_all('div', class_='thumb col4')
    
    for restaurant in restaurants[:20]:
        name = restaurant.find('h2', class_='etablissement-title').text.strip()
        
        address = restaurant.find('div', class_='etablissement-info').text.strip()
        
        rating = restaurant.find('span', class_='note-avis').text.strip()
        
        print(f"Nom: {name}")
        print(f"Adresse: {address}")
        print(f"Note moyenne: {rating}")
        print(response.content)
        print(restaurants)
        print()
else:
    print('La requête a échoué.')
```
